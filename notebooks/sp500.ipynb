```python
import pandas as pd
import bs4 as bs
import urllib.request
import yfinance as yf
from datetime import datetime
```

```python
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

source = urllib.request.urlopen(url).read()

soup = bs.BeautifulSoup(source, 'lxml')

table = soup.find('table', {'id': 'constituents'})

sp500_ticker = []
sp500_name = []
sp500_gics_sector = []
sp500_gics_sub_industry = []

for row in table.find_all('tr')[1:]:
    cells = row.find_all('td')[:4]
    sp500_ticker.append(cells[0].text.strip())
    sp500_name.append(cells[1].text.strip())
    sp500_gics_sector.append(cells[2].text.strip())
    sp500_gics_sub_industry.append(cells[3].text.strip())

print(len(sp500_ticker))
print(len(sp500_name))
print(len(sp500_gics_sector))
print(len(sp500_gics_sub_industry))
```

```python
data = {'ticker': sp500_ticker, 'name': sp500_name, 'sector': sp500_gics_sector, 'industry': sp500_gics_sub_industry}

spx = pd.DataFrame(data=data)
spx
```

```python
start_date = '1990-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')

spx_price = pd.DataFrame(columns=sp500_ticker)

for ticker in sp500_ticker:
    spx_price[ticker] = yf.download(ticker, start_date, end_date)['Adj Close']
```

```python
spx_price_1y = spx_price.iloc[len(spx_price)-252:len(spx_price)+1]
spx_price_1y
```

```python
(spx_price_1y.iloc[251] / spx_price_1y.iloc[0]).sort_values(ascending=False)[:20]
```
